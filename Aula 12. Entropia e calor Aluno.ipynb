```python
def importar_txt(caminho):
    import re
    
    #abrindo o arquivo txt
    r=open(caminho, 'r')
    linhas=r.readlines()
    
    #dic para armazenar os pontos
    points={}
    points['x'] , points['y'] = [], []
    
    #varrendo as linhas
    for i in linhas:
        vals = {}
        vals['x'] = re.findall(r'-?[0-9]+\.?[0-9]*e?\-?[0-9]*(?=\,)',i)[0]
        vals['y'] = re.findall(r'(?<=\,)-?[0-9]+\.?[0-9]*e?\-?[0-9]*',i)[0]

        #procurando por fatores de ordem de magnitude
        for j in ('x', 'y'):
            #print('text val: ', vals[j])
            base_val = re.findall(r'-?[0-9]+\.?[0-9]*', vals[j])
            #print('base val', base_val)            
            factor = re.findall(r'(?<=[0-9]e)\-?[0-9]+', vals[j])
            #print('factor', factor)

            if len(factor) > 0:
                #caso seja negativo
                if factor[0][0] == '-':
                    raw_val = float(base_val[0]) / (10**int(factor[0][1:]))
                #caso seja positivo
                else:
                    print(float(base_val[0]))
                    raw_val = float(base_val[0]) * (10**int(factor[0]))
            else:
                raw_val = float(base_val[0])
        
            rounded_val = round(raw_val, 10)
            #print(f'{j} val to add to the list: ', rounded_val)
            points[j].append(rounded_val)
    
    return points['x'], points['y']
```

```python
def area_sob_curva(*args, n_bars = 1, return_total_area = True):
    from bokeh.plotting import show
    import numpy as np
    
    p = grafico_pontos(*args, return_figure = True)
    
    lista_cores = ['black', 'navy', 'red', 'green', 'orange', 'violet']
    
    results = {}
    results['bar_pos'] = []
    results['area'] = []
    
    if len(args)%2 == 0:
        counter_cores = 0
        for i in range(len(args)):
            area = 0
            if i%2 == 0:
                X = args[i]
                Y = args[i + 1]
                edges = np.array([], dtype=float)
                top = np.array([], dtype=float)

                for j in range(len(X[ : -1])):
                    #x
                    x_low_lim , x_up_lim = X[j] , X[j + 1]
                    delta_x = x_up_lim - x_low_lim
                    dx = abs(delta_x / n_bars)
                    if x_up_lim < x_low_lim:
                        dx = dx * -1
                    l_edges = np.arange(x_low_lim, x_up_lim, dx)
                    edges = np.hstack([edges, l_edges])

                    #y
                    y_low_lim , y_up_lim = Y[j] , Y[j + 1]
                    delta_y = y_up_lim - y_low_lim
                    dy = abs(delta_y / n_bars)
                    if y_up_lim < y_low_lim:
                        dy = dy * -1
                    if y_low_lim == y_up_lim:
                        l_top = np.array([y_low_lim]*n_bars)
                        top = np.hstack([top, l_top])
                    else:
                        l_top = np.arange(y_low_lim, y_up_lim, dy)
                        top = (np.hstack([top, l_top]))
            
                #adicionado o último ponto em x
                edges = np.r_[edges, edges[-1] + dx]                
                p.quad(left=edges[:-1], right=edges[1:], top=top, bottom=0, color=lista_cores[counter_cores], alpha = 0.3)
                counter_cores += 1
                
                #cálculo area
                bar_pos = []
                for i in range(len(edges[:-1])):
                    bar_min = edges[i]
                    bar_max = edges[i + 1]
                    bar_pos.append( (bar_max + bar_min) / 2 )
                    
                results['bar_pos'] = bar_pos
                results['area'] = list(top*dx)
    show(p)
    if return_total_area is True:
        return sum(results['area'])
    else:
        return results
```

<h2>Descrição da Atividade</h2>
<p style='text-align: justify'>
    A capacidade calorífica de um sistema com volume constante pode ser calculada através da relação:</p>

<p style='text-align: center'>
    C<sub>V</sub> = ( ∂U / ∂T )<sub>N, V</sub></p>
    
<p style='text-align: justify'>
    Portanto, precisaríamos ter uma função <i>U(T)</i> para calcular C<sub>V</sub>. Fica a pergunta, como chegar em uma função <i>U(T)</i>? Para isso teríamos que:</p>
    <ol>
        <li>Usar a mecânica quântica e análise combinatória para encontrar uma expressão da multiplicidade (Ω) em função de <i>U, V, N</i> e outras variáveis relevantes.</li>
        <li>Encontrar o logaritmo de Ω para calcular a entropia (<i>S</i>)</li>
        <li>Diferenciar (encontrar a derivada de) <i>S</i> em relação à <i>U</i> para encontrar a temperatura.</li>
        <li>Encontrar a função <i>U(T)</i></li>
        <li>Finalmente diferenciar <i>U(T)</i> em <i>T</i> para encontrar C<sub>V</sub></li>
    </ol>

<p style='text-align: justify'>
    É possível fazer isso tudo? Na verdade é. Inclusive fizemos todos esses passos para alguns sistemas simples como sólidos de Einstein acoplados e gases ideias. No caso dos gases ideais, já temos um função de <i>U(T)</i> desde do início do nosso curso. Podemos inclusive encontrar facilmente C<sub>V</sub> para um número N de partículas de gases ideias com:</p>
    
<p style='text-align: center'>
    <i>U(T)</i> = N (3/2) kT</p>

<p style='text-align: center'>
    C<sub>V</sub> = ( ∂U / ∂T )<sub>N</sub> = N (3/2) k</p>

<p style='text-align: justify'>
    Para 1 mol de gás ideal, temos <i>C<sub>V</sub></i> = 3/2 R, como já foi usado em outras atividades anteriores. Entretanto, só conseguimos usar essa abordagem passo a passo para poucos sistemas físicos. Na grande maioria das vezes isso não será possível. Como fazer então? Mesmo que não possamos escrever uma fórmula para a entropia de um sistema, nós ainda podemos medí-la. Se fornecermos um pouco de calor <i>Q</i> para um sistema mantendo o seu volume constante e não permitindo que haja nenhuma forma de trabalho, podemos usar a relação ( ∂S / ∂U ) = ( dS / dU ) = ( 1 / T ). Isso pois, quando todas as outras variáveis do sistema estão fixas, S passa a variar somente em função da variação da energia interna (<i>U</i>). Nesse caso, não precisamos da derivada parcial pois temos S(U), ou seja, uma função de uma variável. Assim:<p>
        
<p style='text-align: center'>
    dS = ( dU / T ) = ( Q / T ) = ( (C<sub>V</sub> dT) / T ) </p>
    
<p style='text-align: justify'>
    Historicamente, a relação dS = ( Q / T ) representa a <b>definição original de entropia</b> apresentada por Rudolf Clausius em 1865. Clausius definiu entropia como a <i>coisa</i> em um sistema que aumenta por um fator ( Q / T ) quando <i>Q</i> entra no sistema a temperatura <i>T</i>.</p>

<h2>Observação</h2>
<ol>
    <li>Em baixas temperaturas, o C<sub>V,mol</sub> (capacidade calorífica molar a volume constante) do alumínio pode ser calculado através de uma função determinada empiricamente:</li>
    <dl>
        <dt>C<sub>V</sub>(T) = a.T + b.T<sup>3</sup>, onde a = 0.00135 J/K<sup>2</sup> e b = 2.48x10<sup>-5</sup> J/K<sup>4</sup></dt>
    </dl>
    <li>Para grandes variações de temperatura, o valor de C<sub>V</sub>(T) tem que ser medido experimentalmente. Os dados presentes no arquivo <b>Aula 12.         </ul>
    <li>Para T = 0, a relação ( Q / T ) é indefinida. O que significa isso? Que não temos um valor definido para entropia. Como isso não é possível de acontecer, uma vez que S ≥ 0, não podemos calcular a entropia para T = 0 usando essa fórmula. Ao invez disso, assumimos uma aproximaçõa que C<sub>V,mol</sub> → 0 quanto T → 0, e S = 0 quando T = 0. Essa afirmação é chamada de <b>terceira lei da termodinâmica</b>. Entretanto, na prática S(0) pode ser diferente de zero. Como um exemplo, temos o fato que em alguns cristais é possível mudar a orientação das moléculas com muito pouca energia. Dessa forma, quando esses cristais atingem T = 0, teríamos uma certa orientação randômica das moléculas, e portanto, não somente 1 arranjo molecular (o que seria equivalente a Ω = 1). Chamamos isso de <b>entropia residual</b>.</li>

```python
def func_Cv(T):
    a = 0.00135
    b = 2.48*10**-5
    return a + b*T**2

def func_delta_S(X, M, N_mol):
    delta_S = X*M/N_mol
    return delta_S
```

```python
def func_lista_Cv(Ti, Tf, p):
    lista_Cv = []
    for i in range(Ti, Tf+1, p):
        tempo = func_Cv(i)
        lista_Cv.append(tempo)
    return lista_Cv

def func_lista_temp(Ti, Tf, p):
    lista_T = []
    for i in range(Ti, Tf+1, p):
        lista_T.append(i)
    return lista_T
```
